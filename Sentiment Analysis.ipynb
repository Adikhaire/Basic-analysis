```
reviews_train = []
for line in open('full_train.txt', 'r'):
  reviews_train.append(line.strip())

reviews_test = []
for line in open('full_test.txt', 'r'):
  reviews_test.append(line.strip())

```

```
import re

REPLACE_NOSPACE = re.compile("[.,:'?:()\"\[\]!]")

REPLACE_WITH_SPACE = re.compile("(<br\s*/><br\s*/>)|(\-)|(\/)")

def preprocess_reviews(reviews):
  reviews = [REPLACE_NOSPACE.sub('', line.lower()) for line in reviews]
  reviews = [REPLACE_NOSPACE.sub(' ', line) for line in reviews]

  return reviews

reviews_train = preprocess_reviews(reviews_train)

reviews_test = preprocess_reviews(reviews_test)
```

```
from sklearn.feature_extraction.text import CountVectorizer

cv = CountVectorizer(binary=True)
cv.fit(reviews_train)
X_train = cv.fit_transform(reviews_train)
X_test = cv.transform(reviews_test)
#cv.get_feature_names()
```

```

from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split


target = [1 if i > 12500 else 0 for i in range(25000)]

X_train_sub, Y_train_sub, X_tar, Y_tar = train_test_split(X_train, target, train_size=0.75)

for c in [0.01, 0.05, 0.25, 0.5, 1]:

  lr =  LogisticRegression(C=c)
  lr.fit(X_train_sub, X_tar)
  print("Accuracy for c %s: %s" % (c, accuracy_score(Y_tar, lr.predict(Y_train_sub))))


```
