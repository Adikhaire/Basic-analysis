```
import requests
from bs4 import BeautifulSoup
```

```
#url to download the table from the website
url = 'https://www.worldometers.info/coronavirus/'
headers = {'User-Agent': 'Chrome/80.0.3987.149'}
```

```
response = requests.get(url, headers = headers)
soup = BeautifulSoup(response.content, 'html.parser')
stats_table = soup.find_all('table', class_ = "main_table_countries")
stats_table = stats_table[0]
```

```
with open('coronavirus.txt', mode='w') as r:
  for row in stats_table.find_all('tr'):
    for cell in row.find_all('th'):
      r.write(cell.text + ';;')
  
  for row in stats_table.find_all('tr'):
    for cell in row.find_all('td'):
      r.write(cell.text.strip() + ';;' )
    r.write('\r\n')
#from google.colab import files
#files.download('coronavirus.txt')
```

```
import csv
with open('coronavirus.txt', mode='r') as file:
  stripped = (line.strip() for line in file)
  lines = (line.split(';;') for line in stripped if line)
  with open('log.csv', 'w') as out_file:
        writer = csv.writer(out_file)
        writer.writerows(lines)
#files.download('log.csv')
```

```
import pandas as pd
data = pd.read_csv('log.csv')
data = data.iloc[:len(data)-1, :data.shape[1]-1]
data.head()
```

```
import folium
import pandas as pd

country_geo = 'world-countries.json'
world = data.loc[:, ['Country,Other', 'TotalCases']]
world['Country,Other'] = world['Country,Other'].astype(str)
world['TotalCases'] = world['TotalCases'].str.replace(',', '')
world['TotalCases'] = world['TotalCases'].astype(int)
hist_indicator = 'Total Cases'
```

```
map = folium.Map(location=[100, 0], zoom_start=3,  tiles='OpenStreetMap')

map.choropleth(geo_data=country_geo, data=world, columns = ['Country,Other', 'TotalCases'], key_on='feature.properties.name', 
               fill_color='YlOrRd', fill_opacity=0.6, line_opacity=0.2,  legend_name=hist_indicator)
map
#
#map.save('plot_data.html')
# Import the Folium interactive html file

#files.download('plot_data.html')
#from IPython.display import HTML
#file = plot_data.html
#HTML('<iframe src=plot_data.html width=700 height=450></iframe>')
```
