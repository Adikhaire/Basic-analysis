```
import requests

resp = requests.get('https://jsonplaceholder.typicode.com/todos/1')
if resp.status_code != 200:
  raise ApiError('Get /tasks/ {}'.format(resp.status_code))
count = 0
for toDoItem in resp:
  print(toDoItem)
  print(resp.json())
  count = count + 1
  if count > 10:
    break
```

```
# How the post the data through an API work
while True:
  try:
      dividend = int(input("Enter the dividend: "))
      divisor = int(input("Enter the divisor: "))
      if dividend < 10:
        raise ValueError("Dividend is less than 10")
      print("%d %d = %f", (dividend, divisor, (dividend/ divisor)))
  except ValueError as err:
    print("Hey that wasn't a number", err)
  except ZeroDivisionError:
    print("Cannot divide by zero")
  else:
    break
```

```
def printKwargs(**person):
  for k, v in person.items():
    print(k, v)

myDict = {"name":"John", "lastname":"Smith"}
printKwargs(**myDict)
```

```
# we define a decorator
def log(original_function):
    def new_function(*args, **kwargs):
        with open("log.txt", "w") as logfile:
            logfile.write("Function '%s' called with positional arguments %s and keyword arguments %s.\n" % (original_function.__name__, args, kwargs))

        return original_function(*args, **kwargs)

    return new_function

# here is a function to decorate
def my_function(message):
    print("Hello")

# and here is how we decorate it
my_function = log(my_function)
```
